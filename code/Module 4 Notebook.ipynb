## Disease Background:
* Cancer hallmark focus: evading immune destruction

* Overview of hallmark: 
    * One way that cancer can spread is by evading the immune system. In a healthy patient, peptide antigens from cancer cells are presented to T cells by major histocompatibility complex (MHC) class I. The T cells then recognize this, which triggers an immune response causing lysosome-related organelles to be released. These organelles have a protein called perforin, which can cause the formation of pores in cell membranes, and Granzyme B which induces apoptosis of the cancer cell. These mechanisms work together to effectivly fight off cancer.
    * In a disease state, cancerous tumors can block immune cells through the activity of cancer-associated fibroblasts which create a physical barrier through fibrosis. Additionally, they can suppress chemokines which are responsible for recruiting immune cells, or they can produce the chemokine CXCL12 which recruits immune cells in a different location, leading them away from the tumor.
    * Once T cells enter the tumor microenvironment, cancer cells can evade them through the supression of cancer-specific peptide antigens in which the T cells are targeting. Furthermore, a reduction in interferon responsiveness enhances this strategy, as interferon responsiveness upregulates antigen presentation.
    * Even when T cells recognize the peptide antigens, the cytotoxicity of T cells can be regulated by an increased expression of ligands for immune inhibitory receptors, which hinders the function of T cells.
        * https://blog.cellsignal.com/hallmarks-of-cancer-avoiding-immune-destruction

* Oncogenes associated with immune evasion (you can point to a CSV with the list of genes you will study, or point to a citation with a gene signature of interest)
    * MYC: Increases expression of PD-L1
    * EGFR: Increases expression of PD-L1
    * PIK3CA: Reduces recruitment of T cells and increases inhibitory myeloid cells
    * BRAF: Increases expression of PD-L1
    * CTNNB1: Decreases dendritic cell recruitment
    * STAT3: Increases expression of PD-L1, promotes immunosuppressive cytokines
* Tumor suppressor genes associated with immune evasion:
    * PTEN: Decreases infiltration of T cells
    * BRCA1 and BRCA2: Reduces immune responses
    * TP53: Causes an immunosuppressive environment
    * STK11: Hinders infiltration of immune cells
    * RB1: Decreases recruitment of immune cells
    * SMAD4: Increases regulatory T cells and MDSCs
    * APC: Decreases infiltration of T cells
    * ATM: Decreases immune response to DNA damage

When an entire section of the lung is removed, this is called a lobectomy, and when a whole lung is removed, this is a pneumonectomy.

    * Stages of Non-Small Cell Lung Cancer:

There may be several tumors in the same lung lobe, which could have spread to the trachea, carina, esophagus, breastbone, diaphragm, heart, aorta or vena cava, or nerves that control the larynx.

### Methods

The machine learning technique: logistic regression

Logistic regression is a type of classification model, which we chose because we are using gene expression associated with cancer cell immune evasion to predict stages of lung cancer. This method optimizes logistic loss, which means that it calculates the probability that a sample fits under a certain label and compares that to the actual labels. The model decides when it's accurate enough by changing its parameters until its cost function reaches a minimum, meaning it has reduced uncertainty and improved accuracy to the best of its ability. Specifically, it aims to reduce both in-sample and out-of-sample error. To reduce in-sample error, the model must be well-fit to the training data, but to mitigate out-of-sample error, the model must not be overfit to the training data to the point where it cannot be generalized outside of this data.

PCA:

We performed a Principal Component Analysis (PCA) to reduce the dimensionality of our 12-gene immune evasion hallmark set. This allows us to visualize complex genetic patterns in two dimensions. We generated two separate plots: one colored by Cancer Type to observe biological variability between LUAD and LUSC, and a second colored by Tumor Stage to determine if immune evasion signatures correlate with clinical severity

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Assume df_clean is the DataFrame with gene expression and metadata
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_clean[my_genes])

# Perform PCA to reduce to 2 principal components
pca = PCA(n_components=2)
pcs = pca.fit_transform(X_scaled)

# Create a DataFrame for plotting
plot_df = pd.DataFrame(pcs, columns=['PC1', 'PC2'])
plot_df['Type'] = df_clean['cancer_type'].values
plot_df['Stage'] = df_clean['ajcc_pathologic_tumor_stage'].values

# Simplify stage labels to I, II, III, IV
stage_map = {
    'Stage I': 'Stage I', 'Stage IA': 'Stage I', 'Stage IB': 'Stage I',
    'Stage II': 'Stage II', 'Stage IIA': 'Stage II', 'Stage IIB': 'Stage II',
    'Stage III': 'Stage III', 'Stage IIIA': 'Stage III', 'Stage IIIB': 'Stage III',
    'Stage IV': 'Stage IV', 'Stage IVA': 'Stage IV', 'Stage IVB': 'Stage IV'
}
plot_df['Simplified_Stage'] = plot_df['Stage'].map(stage_map)
plot_df = plot_df.dropna(subset=['Simplified_Stage'])

# Plot PCA results
plt.figure(figsize=(16, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(data=plot_df, x='PC1', y='PC2', hue='Type', s=70, alpha=0.7)
plt.title('PCA: Cancer Type')

plt.subplot(1, 2, 2)
sns.scatterplot(data=plot_df, x='PC1', y='PC2', hue='Simplified_Stage', 
                hue_order=['Stage I', 'Stage II', 'Stage III', 'Stage IV'],
                palette='viridis', s=70, alpha=0.7)
plt.title('PCA: Cancer Severity (Stages I-IV)')

plt.tight_layout()
plt.show()

```

Analysis of PCA: The PCA colored by cancer type shows a slight horizontal separation, with LUSC samples leaning toward the left and LUAD samples leaning toward the right along PC1. This suggests the immune evasion hallmark has distinct expression profiles for each subtype. However, the PCA colored by severity (Stage) shows no clear clustering or trend. This indicates that for this gene set, immune evasion signatures do not necessarily intensify or shift linearly as the cancer progresses from Stage I to Stage IV.

K-Means:

We used K-Means clustering to see if our 12 genes would naturally group the patients together without using any labels like cancer type or stage. We set the algorithm to find three clusters to identify different biological patterns in the data. By running this on our scaled gene expression values and plotting it over our PCA, we can see if there are specific groups of patients that share a similar immune evasion profile. This helps us figure out if the genetic data is actually grouping by something other than the clinical categories we already have.

```python
from sklearn.cluster import KMeans

# Fixes the length error by matching data to the 151 valid samples
X_final = X_scaled[plot_df.index]

# Perform K-Means clustering
km = KMeans(n_clusters=3, n_init='auto', random_state=42)
plot_df['Cluster'] = km.fit_predict(X_final).astype(str)

# Plot K-Means clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=plot_df, x='PC1', y='PC2', hue='Cluster', palette='Set1', s=70)
plt.title('Unsupervised K-Means Clusters on Immune Evasion Data')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

Analysis of K-Means: 

The K-Means clustering shows that the data splits into three distinct regions that don't follow a simple left-to-right line. Red occupies the top left, green is in the bottom middle, and blue takes the top right, forming a sort of triangular arrangement. This layout suggests that PC1 and PC2 are equally important in defining these groups. Since patients from every cancer stage and both types are mixed into these three specific areas, it may support the idea that a tumor's immune evasion signature isn't tied to how far the cancer has spread or whether it is LUAD or LUSC. Instead, these clusters represent three different biological states of immune activity that can exist at any stage of the disease.

### Training Data Model:

```python
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import joblib


# Load data
meta = pd.read_csv(r'data\TRAINING_SET_GSE62944_metadata.csv')
expr = pd.read_csv(r'data\TRAINING_SET_GSE62944_subsample_log2TPM.csv', index_col=0)

# Filter LUAD + LUSC
meta = meta[meta['cancer_type'].isin(['LUAD', 'LUSC'])]

# transpose expression matrix (samples as rows)
expr = expr.T

# Genes
my_genes = [
    'CD274', 'CTLA4', 'LAG3', 'HLA-A', 'B2M', 'STAT3',
    'TGFB1', 'MYC', 'EGFR', 'PIK3CA', 'BRAF', 'CTNNB1',
    'PTEN', 'TP53', 'STK11', 'RB1', 'SMAD4', 'APC', 'ATM'
]

# Build dataset
X_data = expr[my_genes]
df = X_data.join(meta.set_index('sample'), how='inner')

# Clean stage labels
df = df.dropna(subset=['ajcc_pathologic_tumor_stage'])

# Simplify stage labels to I, II, III, IV
def simplify_stage(stage):
    if pd.isna(stage):
        return None
    stage = str(stage).upper()

    if "I" in stage and "II" not in stage and "III" not in stage and "IV" not in stage:
        return "I"
    elif "II" in stage and "III" not in stage and "IV" not in stage:
        return "II"
    elif "III" in stage and "IV" not in stage:
        return "III"
    elif "IV" in stage:
        return "IV"
    else:
        return None

df["stage"] = df["ajcc_pathologic_tumor_stage"].apply(simplify_stage)
df = df.dropna(subset=["stage"])

# Features and target
X = df[my_genes]
y = df["stage"]

# Model
model = LogisticRegression(max_iter=2000)
model.fit(X, y)

# Training accuracy
y_pred = model.predict(X)
acc = accuracy_score(y, y_pred)

print(f"In-sample Accuracy (Stage Prediction): {acc:.2f}")

# Save model
joblib.dump(model, "lung_stage_model.pkl")
```

## Conclusions and Ethical Implications: 

The integration of Principal Component Analysis, K-Means clustering, and Logistic Regression models demonstrates that while immune evasion gene expression is a fundamental hallmark of lung cancer, it is not a reliable predictor of clinical pathologic stage. The PCA and K-Means results suggest that these genetic signatures do not follow a linear progression with tumor severity but rather exist in varying stages at varying levels.

The lack of a clear specific signal for a given stage led to the machine learning models to be deceptive in accuracy. The heavy class imbalance of 81 percent early-stage samples led the algorithm to default to Early predictions. Consequently, the binary model practically failed in identifying late-stage cases, correctly predicting only 1 out of 29. In a diagnostic setting, such high false-negative rates for advanced disease would lead to dangerous under-treatment and inequitable care, proving that these markers must be balanced with other hallmarks or clinical data to be useful for severity prediction. 

The greatest ethical risk of using this model would be late-stage false-negatives. Therefore, it is important that any clinical diagnosis be supported by validated information that is more consistent in stage classigifcation than our model. Lastly, it is always important to understand cultural norms in giving diagnoses; the delivery of these results requires a balance of medical accuracy and empathetic communication that respects the patient's background.

## Limitations and Future Work: 

Based on the conclusions from the analysis, solely immune suppressing gene expression is not enough to determine stage accurately. In the future, it would be helpful to look at another indicator, like tissue invasion and metastasis gene markers, along with immune evasion to find stronger predictors of severity. Our primary limitation was the class imbalance in the TCGA dataset where 81 percent of validation samples were early stage, leaving only 29 percent late stage cases for the model to learn from.

To address this in future work, a method like SMOTE, or Synthetic Minority Over-sampling Technique, could be used to balance the dataset. SMOTE is a statistical technique that creates new, synthetic examples of the minority class, which in our case is late stage cancer, by interpolating between existing data points rather than just duplicating them. This prevents the model from falling into an accuracy trap and forces it to learn the specific genetic features of advanced disease.
